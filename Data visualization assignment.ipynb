```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets 
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
X=iris.drop('species',axis=1)
y=iris['species']
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
X_new=pca.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
X_new
```

```python
y
```

```python
label_mapping = {l: i for i, l in enumerate(labelencoder_y.classes_)}
label_mapping
```

```python
target=list(label_mapping.keys())
```

```python
target
```

```python
fig=plt.figure(1,figsize=(8,6))
ax=Axes3D(fig)
ax.scatter(X_new[:,0],X_new[:,1],X_new[:,2],c=y,cmap=plt.cm.rainbow, edgecolor='k')
ax.set_title('PCA_Iris_Data_Set')
ax.set_xlabel('1st_component')
ax.set_ylabel('2nd_component')
ax.set_zlabel('3rd_component')
ax.legend(target)
plt.show()
```
